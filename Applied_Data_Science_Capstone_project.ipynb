**Week 3**

Scraping the Wikipedia site

```python
#Using BeautifulSoup to scrape the Wikipedia site
from bs4 import BeautifulSoup
import requests
```

```python
source=requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text #scraping the site
```

```python
soup=BeautifulSoup(source, 'lxml')#Using BeautifulSoup to create a list of the site and the lxml
#print(soup.prettify())#Using prettify to get the indentation
```

```python
data =[] #creating an empty list
table_body=soup.find('tbody')#put all the content of the table on the site into a list
print(table_body)
```

```python
rows = table_body.find_all('tr')#create rows from the list
for row in rows: #add the rows into the data-list
    cols = row.find_all('td')
    cols = [ele.text.strip() for ele in cols]
    data.append([ele for ele in cols if ele]) # Get rid of empty values
```

**Week 3** 

Creating the dataframe

```python
#print(data)
df_neighbourhoods=pd.DataFrame(data)#transform into a dataframe
df_neighbourhoods
```

```python
df_neighbourhoods.columns =['Postcode','Borough','Neighbourhood']#change the headers
```

```python
df_neighbourhoods
```

```python
df_neighbourhoods=df_neighbourhoods.dropna()#remove empty values
```

```python
indexNotAssigned=df_neighbourhoods[df_neighbourhoods['Borough']== 'Not assigned'].index #Get indicies of rows with Not assigned in the Borough column
df_neighbourhoods.drop(indexNotAssigned, inplace=True)#dropping line items without Borough
df_neighbourhoods.head()
```

```python
df_neighbourhoods
```

```python
df_neighbourhoods.Neighbourhood.replace('Not assigned',df_neighbourhoods.Borough )
df_neighbourhoods
```

```python
df_output=df_neighbourhoods.groupby(['Postcode','Borough'])['Neighbourhood'].agg(', '.join).reset_index()#Grouping by Postcode and Borough to get the correct look according to instructions, adding the Neighbourhood column joining the rows with the "," delimiter and restting the index
df_output.head()
```

**Clustering**

```python
from sklearn.cluster import KMeans
import matplotlib.cm as cm
import matplotlib.colors as colors
print('Libraries imported.')
k_means = KMeans(init="k-means++", n_clusters=6, n_init=12) #Setting up k-means
```

```python
Toronto_crime_clustering=Toronto_latlong_crime.drop('Borough',1) #dropping the non numerical columns
```

```python
Toronto_crime_clustering=Toronto_crime_clustering.drop('Neighbourhood',1)#dropping the non numerical columns
```

```python
Toronto_crime_clustering=Toronto_crime_clustering.fillna(0) #Removing NaN
k_means.fit(Toronto_crime_clustering)
```

```python
k_means_labels = k_means.labels_ #saving the labels of each point
k_means_labels
```

```python
k_means_cluster_centers = k_means.cluster_centers_ #saving the coordinates of the cluster centers/centroids
k_means_cluster_centers
```

```python
Toronto_latlong_crime['Clustering']=k_means_labels
Toronto_latlong_crime=Toronto_latlong_crime.fillna(0)
Toronto_latlong_crime
```

```python
n_clusters=6 #from the initialization of Kmeans
# creating the clustered map
clustered_map = folium.Map(location=[latitude, longitude], zoom_start=12)

# set color scheme for the clusters
x = np.arange(n_clusters)
ys = [i + x + (i*x)**2 for i in range(n_clusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(Toronto_latlong_crime['Latitude'], Toronto_latlong_crime['Longitude'], Toronto_latlong_crime['Neighbourhood'], Toronto_latlong_crime['Clustering']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(clustered_map)
       
clustered_map
```
